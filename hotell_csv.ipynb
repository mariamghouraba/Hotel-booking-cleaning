```python
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
df['agent'].fillna(df['agent'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
df['adr']= np.where(df['adr']<0,0,df['adr'])
```

```python
df['adr']= np.where(df['adr']>510,510,df['adr'])
```

```python
plt.figure(figsize=(15,15))
sns.histplot(data=df,x='adr')
```
